```python
# multiple lines 
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])
y3 = np.array([1,1,1,1])
y4 = np.array([1,2,3,4])
plt.plot(y1, marker='o')
plt.plot(y2, marker='o')
plt.plot(y3, marker='o')
plt.plot(y4, marker='o')
plt.show()
```

```python
x1 = np.array([0,1,2,4,5])
y1 = np.array([3,1,8,6,5])
x2 = np.array([0,1,2,3,4])
y2 = np.array([6,3,2,10,4])

font1 = {'family': 'serif', 'color': 'blue', 'size':20}

plt.plot(x1, y1, x2, y2, marker='o')
plt.xlabel('x axis title')
plt.ylabel('y axis title')
plt.title('Data Dashboard', loc='center', fontdict=font1)
plt.grid(axis = 'y', color='#0c400b', linestyle='--', linewidth = 0.5) #define a grid line
plt.show()
```

```python
#subplot
#plot1
x = np.array([0,1,3,4])
y = np.array([3,8,5,6])
plt.subplot(1, 2, 1) # row , column, plot index
plt.plot(x,y, marker='o')

#plot2
x = np.array([0,1,2,3])
y = np.array([5,10,7,9])
plt.subplot(1, 2, 2) # row , column, plot index
plt.plot(x,y, marker='o')

plt.show()
```

```python
#Draw 6 plot
#plot1
x = np.array([0,1,2,3])
y = np.array([3,6,5,7])

plt.subplot(2,3,1)
plt.plot(x, y)
plt.title("SALES")

#plot2
x = np.array([0,1,2,3])
y = np.array([3,5,10,7])

plt.subplot(2,3,2)
plt.plot(x, y)
plt.title("INCOME")

#plot3
x = np.array([0,1,2,3])
y = np.array([1,6,11,5])

plt.subplot(2,3,3)
plt.plot(x, y)
plt.title("USAGE")

#plot4
x = np.array([0,1,2,3])
y = np.array([0,6,1,9])

plt.subplot(2,3,4)
plt.plot(x, y)
plt.title("TIME")

#plot5
x = np.array([0,1,2,3])
y = np.array([1,0,1,9])

plt.subplot(2,3,5)
plt.plot(x, y, marker='o')
plt.title("BILL")

#plot6
x = np.array([0,1,2,3])
y = np.array([1,0,1,0])

plt.subplot(2,3,6)
plt.plot(x, y)
plt.title("GROW")

plt.suptitle('MY STORE')

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 200])

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
